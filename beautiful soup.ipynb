```python
def make_soup(url):
    thepage=urllib.request.urlopen(url)
    soupdata=soup(thepage,"html.parser")
    return soupdata
playerdatasaved=""
for eachurl in myurls : 
    soup1= make_soup(eachurl)
    for record in soup1.findAll("div",{"class":"genTable"})[:1]:
        if record==['\r', '\t\t\t\tThere is no data for this month, please select a prior month.\r', '\t\t\t']:
            continue
        else:
            playerdata=""
            for data in record.findAll('tr')[1:]:
                playerdata=playerdata+data.text
            playerdatasaved=playerdatasaved+"\n"+playerdata[1:]

```

```python
playerdatasaved2= playerdatasaved.split("\n")
playerdatasaved3= list(filter(None,playerdatasaved2))
playerdatasaved3= np.asarray(playerdatasaved3)
playerdatasaved3=playerdatasaved3.reshape(-1,7)[1:]
df6= pd.DataFrame(playerdatasaved3, columns=("company_name", "symbol"," market", "price", 'shares', 'offer_amount', 'date_priced'))

```
